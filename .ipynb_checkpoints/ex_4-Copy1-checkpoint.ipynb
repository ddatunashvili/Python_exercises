# 4
დაწერე ფუნქცია,რომელიც მიიღებს რიცხვს დააბრუნებს რიცხვს ერთით გაზრდილს ოღონდ არ გამოიყენოთ  მათემატიკური ოპერატორები

* ტექსტში ყოველი ასო ერთი ინდექსით გადაწეულია მარჯვნივ 
* შეგიძლიათ გამოიყენოთ index() ფუნქცია
* არ დაგავიწყდეს შედეგის დაბრუნება  (return)
* თუ კოდი არ ეშვება წაშალეთ ან დააკომენტარეთ 

მაგალითი:

increment("ebwje") ➞"david"

increment("fbsui") ➞ "earth"

increment("mjovy") ➞ "linux"


```python

#__start__
def decrypt(text):
    lang = "abcdefghijklmnopqrstuvwxyz"
    result = ""
    
    for i in text:
        result += lang[lang.index(i)+1]
    return result
        
    
decrypt("linux")
        
        

#__end__
```
