# 1

დაწერე ფუნქცია,რომელიც გვიჩვენებს რამდენად მდგრადია რიცხვი

* არ დაგავიწყდეს შედეგის დაბრუნება  (return)
* თუ კოდი არ ეშვება წაშალეთ ან დააკომენტარეთ 





მაგალითი:

persistence(39) ➞ 3    3*9 = 27, 2*7 = 14, 1*4 = 4,  
                                          
persistence(999) ➞ 4   9*9*9 = 729,   7*2*9 = 126, 1*2*6 = 12,  1*2 = 2

persistence(4) ➞ 0     ერთნიშნაა


```python

#__start__

def persistence(n, count = 0):
    n = str(n)
    number = 1
    for i in n:
        number *= int(i)
    count += 1
    if number > 10:
        persistence(number,count)
    else: 
        print(count)
    
persistence(39)

#__end__
```

# 2

დაწერე ფუნქცია,რომელიც მიიღებს ორ სტრინგს და დააფორმატებს შემდეგი ფორმულირებით:

მოკლე სტრინგი+რევერსირებული გრძელი+მოკლე სტრინგი 
(shorter+reverse(longer)+shorter)

* არ დაგავიწყდეს შედეგის დაბრუნება  (return)
* თუ კოდი არ ეშვება წაშალეთ ან დააკომენტარეთ 





მაგალითი:

shorter_reverse_longer("firstt", "abcde") ➞ "abcdettsrifabcde"

shorter_reverse_longer("second", "third") ➞ "thirddnocesthird"

shorter_reverse_longer("grandma", "cat") ➞ "catamdnargcat"




```python
def shorter_reverse_longer(a,b):
    
    if len(a) > len(b):
        short = b
        long = a
    else:
        short = a
        long = b
        
    return short+long[::-1]+short


shorter_reverse_longer("grandma", "cat")
```
