* კლასებს შეუძლიათ უნარები მიიღონ ერთმანეთისგან და შექმნან აბსტრაქტული იერარქია, რომელსაც მემკვიდრეობას ვეძახით.

* აბსტრაქციის აწყობის უამრავი სტილი არსებობს გამომდინარე პროექტიდან თუ სამუშაოდან


````python
class Animal:
    def __init__(self,name):
        # თვისებები:
        self.name = name
    # უნარები:
    def talk(self):
        return f"My name is {self.name}. I am talking"

dog = Animal("Jeka")
cat = Animal("Jessica")
print(dog.talk())
print(cat.talk())

````

<h2 align="center">სტრუქტურა და გამოყენება</h2>

````python
def add(a,b):
    return  a + b

def multiply(a,b):
    return a * b
def minus(a,b):
    return a - b

def math(action, x,y):
    return action(x,y)

print(math(add,1,2))
print(math(multiply,2,2))
print(math(minus,4,2))
````

<h2 align="center">იგივე OOP_ის გამოყენებით  </h2>

````python
class math:
    def add(self,a,b):
        return  a + b

    def multiply(self,a,b):
        return a * b
    
    def minus(self,a,b):
        return a - b

math_object = math()
print(math_object.add(1,2))
print(math_object.multiply(2,2))
print(math_object.minus(4,2))


````
